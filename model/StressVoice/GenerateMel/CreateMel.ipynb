# import necessary libraries

```python
import librosa
import numpy as np
import os
import librosa.display
import matplotlib.pyplot as plt
```

```python
folderList=[]
for path, folders, files in os.walk('./voice_dataset'):
        
        for file in folders:
            folderList.append(os.path.join(path, file))
for singleFolder in folderList:
    image_path=singleFolder.split('./')[1].split('/')[1]
    image_folder='./'+image_path+'/'
    if not os.path.exists(image_path):
        os.makedirs(image_path)
    fileList = []
    for path, folders, files in os.walk(singleFolder):
        for file in files:
            fileList.append(os.path.join(path, file))
    for AUDIO_FILE in fileList:   
        image=AUDIO_FILE.split('.wav')[0].split('./')[1].split('/')[2]
        print(image)
        samples, sample_rate = librosa.load(AUDIO_FILE, sr=None)
        sgram = librosa.stft(samples)
        sgram_mag, _ = librosa.magphase(sgram)
        mel_scale_sgram = librosa.feature.melspectrogram(S=sgram_mag, sr=sample_rate)

        mel_sgram = librosa.amplitude_to_db(mel_scale_sgram, ref=np.min)
        s=librosa.display.specshow(mel_sgram, sr=sample_rate, x_axis='time', y_axis='mel')


        from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
        plt.axis('off')
        fig = plt.Figure()
        canvas = FigureCanvas(fig)
        ax = fig.add_subplot(111)
        p = librosa.display.specshow(mel_sgram, ax=ax)

        fig.savefig(image_folder+image+'.jpg',bbox_inches='tight' )
    print("****** END OF"+singleFolder)
```
